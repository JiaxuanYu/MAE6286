```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 16

nx = 41
L  = 2.0
dx = L / (nx - 1)
nt = 25
dt = 0.02
c  = 1.0

x  = np.linspace(0.0, L, num=nx)
```

```python
u0 = np.ones(nx)
mask = np.where(np.logical_and(x >= 0.5, x <= 1.0))
print(mask)
```

```python
u0[mask] = 2.0
print(u0)
```

```python
plt.figure(figsize=(8, 6))
plt.title('Initial conditions')
plt.xlabel('x')
plt.ylabel('u')
plt.grid()
plt.plot(x, u0, color='b', linestyle='--', linewidth=2)
plt.xlim(0.0, L)
plt.ylim(0.0, 2.5);
```

```python
u = u0.copy()
for n in range(1, nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i - 1])
```

```python
plt.figure(figsize=(8,6))
plt.xlabel('x')
plt.ylabel('u')
plt.grid()
plt.plot(x, u0, label='Initial',color='b', linestyle='--', linewidth=2)
plt.plot(x, u, label='nt = {}'.format(nt), color='r', linestyle='-', linewidth=2)
plt.legend()
plt.xlim(0.0, L)
plt.ylim(0.0, 2.5);
```

```python
nx = 41
L  = 2.0
dx = L / (nx - 1)
nt = 10
dt = 0.02

x = np.linspace(0.0, L, num=nx)
u0 = np.ones(nx)
mask = np.where(np.logical_and(x >= 0.5, x <= 1.0))
u0[mask] = 2.0
```

```python
plt.figure(figsize=(8,6))
plt.title('Initial conditions')
plt.xlabel('x')
plt.ylabel('u')
plt.grid()
plt.plot(x, u0, color='b', linestyle='--', linewidth=2)
plt.xlim(0.0, L)
plt.ylim(0.0, 2.5);
```

```python
u = u0.copy()
for n in range(1, nt):
    u[1:] = u[1:] - dt / dx * u[1:] * (u[1:] - u[:-1])
```

```python
plt.figure(figsize=(4.0, 4.0))
plt.xlabel('x')
plt.ylabel('u')
plt.grid()
plt.plot(x, u0, label='Initial', color='b', linestyle='--', linewidth=2)
plt.plot(x, u, label='nt = {}'.format(nt), color='r', linestyle='-', linewidth=2)
plt.legend()
plt.xlim(0.0, L)
plt.ylim(0.0, 2.5);
```
